In agricultural studies, predicting crop yield accurately is crucial for optimising resource use and improving farm productivity. While factors like historical yield provide some insight into expected future yields, environmental variables such as soil health and water content can play an even more significant role. One such indicator of soil health is microbial OTU (Operational Taxonomic Units) diversity, which measures the variety of microorganisms in the soil, potentially influencing nutrient cycling and plant growth.

However, microbial activity is dependent on soil moisture. Water content can either enhance or limit the effectiveness of soil microbes. By calculating the ratio of OTU diversity to water content, one can normalise the influence of microbial diversity and water availability, possibly revealing a clearer relationship with current yield. This ratio accounts for both microbial activity and soil moisture, enabling one to understand the combined impact of these key variables on crop performance and hopefully leading to more precise yield prediction.

h) Create a function, called *otu_per_water*, that:
*   takes (1) OTU diversity and (2) soil moisture content as inputs
*   calculates and returns the result of dividing the OTU diversity with the soil moisture content

**(3 marks)**


```python
def otu_per_water (otu_diversity,water_content):
  otu_per_water= otu_diversity/water_content
  return otu_per_water
```
